#usa.gov data

This example is taken and adapted from the "Data Analysis for Python" book by Wes MCKinney.
First we load data from URL shortening service bit.ly about the usa.gov domain.


```python
%matplotlib inline
path = 'data/usagov_bitly_data2012-03-16-1331923249.txt'
```

```python
open(path).readline()
```

```python
import json
records = [json.loads(line) for line in open(path)]
```

```python
records[0]
```

```python
records[0]['tz']
```

```python
time_zones =[rec['tz'] for rec in records if 'tz' in rec]
```

```python
time_zones[:10]
```

```python
def  get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts

```

```python
counts = get_counts(time_zones)
counts['America/New_York']
```

```python
len(time_zones)
```

```python
def top_counts(count_dict, n = 10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

#Counting Time Zones with Pandas


```python
from pandas import DataFrame, Series
import pandas as pd; import numpy as np
```

```python
tz_counts = frame['tz'].value_counts()
```

```python
tz_counts = clean_tz.value_counts()
```

```python
cframe= frame[frame.a.notnull()]
```

```python
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows','Not Windows')
```

```python
operating_system[:5]
```

```python
by_tz_os = cframe.groupby(['tz', operating_system])
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
indexer= agg_counts.sum(1).argsort()
```

```python
indexer[:10]
```

```python
count_subset = agg_counts.take(indexer)[-10:]
```

```python
count_subset
```

```python
count_subset.plot(kind='barh',stacked=True)
```

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked= True)
```
